### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
def Attritions(x):
    if x=='Yes':
        return 1
    else:
        if x=='No':
            return 0
```

```python
df['Attrition']=df['Attrition'].apply(Attritions)
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf=pd.get_dummies(df,drop_first=True)
newdf.head()
```

#### Check the shape of our new dataset

```python
newdf.shape
```

#### Print unique values in our new dataframe

```python
for i in newdf:
    print("Number of unique elements in",i,"are",df[i].unique())
```

#### Split the columns into input and target variables (X and y)

```python
X = newdf.drop(columns = ['Attrition'])
y = newdf['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xcolumns = X.columns
```

```python
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns = xcolumns)
```

```python
X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.3, random_state = 100)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print('accuracy_score of our model is :', metrics.accuracy_score(y_test, y_pred))
print('confusion matrix is :', metrics.confusion_matrix(y_test, y_pred))
print('Classification report is :', metrics.classification_report(y_test, y_pred))
print('cross validation is :', cross_val_score(model, X_train,y_train,cv=2).mean())
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
sns.heatmap(metrics.confusion_matrix(y_test, y_pred),annot=True,fmt='d')
```
